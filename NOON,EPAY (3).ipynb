```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df_NOON = pd.read_csv('NOON Historical Data.csv', parse_dates=True)
df_EPAY = pd.read_csv('EPAY Historical Data.csv', parse_dates=True, sep=",")
mergeTable= pd.merge(df_NOON,df_EPAY,how="inner",on='Date')
df = mergeTable[['Price_x','High_x','Price_y']]
df
```

```python
X = df[['Price_x', 'High_x']]
print(X)
y = df['Price_y']
print(y)
```

```python
regression = linear_model.LinearRegression()
regression.fit(X, y)
print(regression.coef_)
print(regression.predict([[0, 0]]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Price_x'], df['High_x'], df['Price_y'], c = 'r', marker = 'o')
xx1, xx2 = np.meshgrid(range(1, 4), range(29, 45))
yy = 21.85*xx1 - 22.30*xx2 + 68.59
ax.plot_surface(xx1, xx2, yy, alpha=0.5)
plt.show()
```
