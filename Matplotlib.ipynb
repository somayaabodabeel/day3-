```python
#Q1
from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x_values = [1, 4, -1, 0, -2]
y_values = [2, 7, 3, -1, -3]
ax.scatter(x_values[0],y_values[0], color = 'r',marker='.')
ax.scatter(x_values[1],y_values[1], color = 'g',marker='*')
ax.scatter(x_values[2],y_values[2], color = 'y',marker='s')
ax.set_xlabel('X AXIS')
ax.set_ylabel('Y AXIS')
ax.set_title ('Scatter Plot')
plt.show()
```

```python
#Q2
#bar graph with six (6) bars – one for each of the categories “A”, “B”, “C”, “D”, “E”, and “F”.
#The heights of the bars should be 7, 3, 12, 4, 2, and 1
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
pets = ['A', 'B', 'C','D','E','F']
nos = [7, 3, 12, 4, 2, 1]
ax.set_xlabel('Categories')
ax.set_ylabel('Heights')
ax.bar(pets, nos)
plt.show()
```

```python
#Q4
import matplotlib.pyplot as plt
x = [15, 25, 25, 30, 5]
labels = ['Asia', 'Europe', 'Africa', 'South America', 'Australia']
colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red']
explode = [0, 0, 0, 0, 0.2]
fig, ax = plt.subplots()
ax.pie(x, labels = labels, colors = colors, autopct='%.0f%%', explode = explode)
ax.set_title('world population by continent')
plt.show()
```
